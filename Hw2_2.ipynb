```python
import pandas as pd

# Правильне посилання: Google Sheets як CSV
sheet_url = "https://docs.google.com/spreadsheets/d/1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ/export?format=csv&gid=1724592523"

# Читання CSV напряму з Google Sheets
df = pd.read_csv(sheet_url)

# Перегляд перших рядків
df.head()
```

```python
# Розмір таблиці
print("Розмір таблиці:", df.shape)
```

```python
# Типи стовпців
print("Типи стовпців:\n", df.dtypes)
```

```python
# Частка пропусків
missing = df.isnull().sum() / len(df)
print("Частка пропусків у кожному стовпці:\n", missing)
```

```python
# Видалення стовпців з пропусками, крім Мова програмування
df = df.loc[:, df.columns[(df.isnull().sum() == 0) | (df.columns == 'Мова програмування')]]

print("Залишились стовпці:", df.columns)
```

```python
# Перевірка пропусків ще раз
print("Частка пропусків після видалення:\n", df.isnull().sum() / len(df))
```

```python
# Видалення всіх рядків з пропусками
df_clean = df.dropna()
print("Новий розмір:", df_clean.shape)
```

```python
# Створення таблицю тільки для Python
python_data = df_clean[df_clean['Мова програмування'] == 'Python']
print("Розмір Python-таблиці:", python_data.shape)
```

```python
# Групування за Посада
grouped = python_data.groupby('Посада')
```

```python
print(df.columns.tolist())
```

```python
# Перегляд всіх назв
print(df.columns)

# заміна крапки на пробіли або підкреслення
df.columns = df.columns.str.replace('.', '_')
print(df.columns)
```

```python
# Мін та макс зарплата
agg_data = grouped.agg({'Зарплата на місяць': ['min', 'max']})
agg_data.reset_index(inplace=True)
print(agg_data)
```

```python
# Функція середньої зарплати
def fill_avg_salary(row):
    return (row[('Зарплата на місяць', 'min')] + row[('Зарплата на місяць', 'max')]) / 2

agg_data['avg'] = agg_data.apply(fill_avg_salary, axis=1)
print(agg_data.head())
```

```python
# Описова статистика для avg
print(agg_data['avg'].describe())
```

```python
# Збереження у CSV
agg_data.to_csv('python_salary_analysis.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Якщо у тебе є:
# agg_data: Посада | min | max | avg

plt.figure(figsize=(12, 6))

# Малюємо стовпчики середньої зарплати
plt.bar(agg_data['Посада'], agg_data['avg'],
        yerr=[agg_data['avg'] - agg_data[('Зарплата на місяць', 'min')],
              agg_data[('Зарплата на місяць', 'max')] - agg_data['avg']],
        capsize=5)

plt.xticks(rotation=90)
plt.xlabel("Посада")
plt.ylabel("Зарплата (мін/середнє/макс)")
plt.title("Середня зарплата з межами мін/макс по посадах")
plt.tight_layout()
plt.show()
```
